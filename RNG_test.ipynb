```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import random
import math
import statistics as stat
```

```python
# Create a 2D lattice 10x10
n=100*100
lattice = np.zeros(n)
unvisited = n
unvisited_arr = []
for i in range(5*n):
    ran = random.randint(0,n-1);
    if(lattice[ran]==0):
        lattice[ran] = ran+1
        unvisited -= 1
    unvisited_arr.append(unvisited)
print(unvisited)
```

```python
plt.scatter(np.arange(len(unvisited_arr)), unvisited_arr, alpha=0.7,edgecolor='black',linewidths=0.5, s=5)
plt.xlabel("Number of Attempts")
plt.ylabel("Number of Unvisited Sites") 
plt.title("Decay of Unvisited Sites")
```
